```python
# load data into dataframe
import pandas as pd
df = pd.read_csv('Spotawheel_case_study (1).csv')
df
```

```python
# create column Brand
df['Brand'] = df['Name'].str.split(' ').str[0] #Separating Brand name from the Name
```

```python
df.info()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

#This heatmap shows correlations between features
plt.figure(figsize=(10,8))
sns.heatmap(df.corr(),annot=True ,cmap="YlGnBu" )
plt.show()
```

```python
# Now take a look also at non-numerical columns

# Impact of Age per Fuel_Type on price 
plt.figure(figsize=(10,7))
plt.title("Price Vs Year VS FuelType")
sns.lineplot(x='Age', y='Price',hue='Fuel_Type',
             data=df)
```

```python
#Impact of Mileage by Transmission type on Price
sns.scatterplot(y='Price', x='Mileage', hue='Transmission', data=df)
```

```python
#Impact of Brand on Price
plt.figure(figsize=(20,15))
sns.set(font_scale=1)
sns.boxplot(x='Price', y='Brand', data=df)
plt.grid()
```

```python
# From the heatmap and the graphs above, we can see that
# Mileage, Age, Engine, Power, Seats,Fuel_Type,Transmission, and Brand, all effect the price.
# Have to convert Fuel_Type, Transmission, and Brand dtype from object to category
# Then we can use categorical data to train
```

```python
# convert columns with dtype object to dtype category, and give different values to each category
df["Fueltype"]= df["Fuel_Type"].astype('category')
df["Fueltypecat"] = df["Fueltype"].cat.codes
```

```python
# convert columns with dtype object to dtype category, and give different values to each category

df["Transmissiontype"]= df["Transmission"].astype('category')
df["Transmissiontypecat"] = df["Transmissiontype"].cat.codes
```

```python
df["Brandtype"]= df["Brand"].astype('category')
df["Brandtypecat"] = df["Brandtype"].cat.codes
```

```python
# split df to be train data and label

# Mileage, Age, Kilometers_Driven, Owner_Type, Seats, Engine, Power,Fuel_Type,Transmission, 
# and Brand effect the price from most negative to most positive. 

x = df[["Mileage", "Kilometers_Driven", "Owner_Type","Fueltypecat", "Transmissiontypecat", "Brandtypecat", "Age", "Engine", "Seats", "Power"]].values
y = df["Price"].values
```

```python
# split train data to be train and val
train_indices = int(0.8 * df.shape[0])
x_train, x_val, y_train, y_val = (
        x[:train_indices],
        x[train_indices:],
        y[:train_indices],
        y[train_indices:],
    )
```

```python
# train my model
import numpy as np
from sklearn.linear_model import LinearRegression
reg = LinearRegression().fit(x_train, y_train)
```

```python
# check the model's performance
reg.score(x_train,y_train)
```

```python
reg.score(x_val, y_val)
```

```python
reg.predict(x_val)
```

```python
# compare with another model that trains with train_test_split data

from sklearn.model_selection import train_test_split
X_train, X_test, Y_train, Y_test = train_test_split(x, y, test_size=0.2, random_state=0)
reg2 = LinearRegression().fit(X_train, Y_train)
reg2.score(X_train, Y_train)
```

```python
print(reg.intercept_)
```

```python
print(reg.coef_)
```

```python
y_pred = reg.predict(x_val)
```

```python
# evaluation of the model, not very accurate, but decent

from sklearn import metrics
print('Mean Absolute Error:', metrics.mean_absolute_error(y_val, y_pred))
print('Mean Squared Error:', metrics.mean_squared_error(y_val, y_pred))
print('Root Mean Squared Error:', np.sqrt(metrics.mean_squared_error(y_val, y_pred)))
```

```python
# prepare evaluation metric
evalua = pd.DataFrame({'Actual': y_val, 'Predicted': y_pred})
evalua
```

```python
# plot evaluation graph

from sklearn.model_selection import cross_val_predict
from sklearn import linear_model
import matplotlib.pyplot as plt

predicted = cross_val_predict(reg, x, y, cv=10)

fig, ax = plt.subplots()
ax.scatter(y, predicted, edgecolors=(0, 0, 0))
ax.plot([y.min(), y.max()], [y.min(), y.max()], 'k--', lw=4)
ax.set_xlabel('Measured')
ax.set_ylabel('Predicted')
plt.show()
```
